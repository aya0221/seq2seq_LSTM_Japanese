# preprocessing (データの前処理)-seq2seq-

## read the data and format the data (データの正規化処理)

```python
import re #正規表現のためにreをインポート

#novels = ["pepper.txt"]
novels = ["pepper.txt", "japanese_corpus1.txt", "japanese_corpus2.txt", "japanese_corpus3.txt", "japanese_corpus4.txt", "japanese_corpus5.txt"]

text = ""
for novel in novels:
    with open("data/"+novel, mode="r", encoding="utf-8") as f:  # ファイルの読み込み
        text_novel = f.read()
    text_novel = re.sub("《[^》]+》", "", text_novel)  # ルビの削除
    text_novel = re.sub("［[^］]+］", "", text_novel)  # 読みの注意の削除
    text_novel = re.sub(r"\\(.+)\=(.+)\\", "", text_novel)  # \文字=文字\の削除
    text_novel = re.sub(r"＜(.+)＞", "", text_novel)  # ＜＞の削除
    text_novel = re.sub("[|]", "", text_novel)  # |の削除
    text_novel = re.sub("ペッパー", "トモット", text_novel)  # |の削除
    text_novel = re.sub(r"\[(.+)\]", "", text_novel)  # \文字=文字\の削除
    text_novel = re.sub("〔[^〕]+〕", "", text_novel)  # 読みの注意の削除
    text_novel = re.sub("[ 　\n「」『』（）｜※＊…]", "", text_novel)  # 全角半角スペース、改行、その他記号の削除
    text += text_novel

print("文字数:", len(text))
print(text)
```

## 漢字→ひらがなに変換

```python
pip install pykakasi
```

```python
from pykakasi import kakasi

seperator = "。"  # 。をセパレータに指定
sentence_list = text.split(seperator)  # セパレーターを使って文章をリストに分割する
sentence_list.pop() # 最後の要素は空の文字列になるので、削除
sentence_list = [x+seperator for x in sentence_list]  # 文章の最後に。を追加

kakasi = kakasi()
kakasi.setMode("J", "H")  # J(漢字) からH(ひらがな)へ
conv = kakasi.getConverter()
for sentence in sentence_list:
    print(sentence)
    print(conv.do(sentence))
    print()
```

set()を使って文字の重複を無くす

```python
kana_text = conv.do(text)  # 全体をひらがなに変換
print(set(kana_text))  # set()で文字の重複をなくし、使用されている文字の一覧を表示
```

## テキストデータの保存
テキストデータを保存し、いつでも使えるようにする。

```python
print(kana_text)
with open("kana_pepper.txt", mode="w", encoding="utf-8") as f:
    f.write(kana_text)
```
